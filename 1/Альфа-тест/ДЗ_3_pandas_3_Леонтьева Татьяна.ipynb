4. ** Мы хотим понять, какие фильмы приносят нам больше выручки — дорогие или дешевые? За уровень, разделяющий фильмы на дорогие/дешевые примите среднее значение по столбцу "цена билета".

```python
# your code
# посчитаем среднюю цену билета
mean_price = df['price_ticket'].mean()
```

```python
# посчитаем выручку по дорогим и дешевым билетам
expensive_tickets = df[df['price_ticket'] > mean_price]['revenue'].sum()   
cheap_tickets = df[df['price_ticket'] < mean_price]['revenue'].sum()
 
print(f"Выручка от продажи дорогих билетов составляет: {expensive_tickets} рублей")
print(f"Выручка от продажи дешевых билетов составляет: {cheap_tickets} рублей")
```

```python
# Сравним выручку от продажи дорогих и дешевых билетов:
if expensive_tickets > cheap_tickets:
    print(f'Выгоднее продавать дорогие билеты')
elif cheap_tickets > expensive_tickets:
     print(f'Выгоднее продавать дешевые билеты')
else:
    print(f'Кинотеатр получил одинаковую выручку за указанный период')  
    
    
    
    
```
